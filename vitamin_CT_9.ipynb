```python
mpg = pd.read_csv("C:/Users/82109/data_setting/data/mpg.csv")
mpg.head()
```

```python
mpg = mpg.sort_values(by='fl')
mpg['fl'].value_counts(ascending = True)
```

```python
mpg1 = pd.DataFrame({'fl':[],'price_fl':[]})
list_fl = mpg['fl']
list_fl = list(list_fl)
mpg1['fl']=list_fl
```

```python
#np.where를 이용 값처리!
mpg1['price_fl'] = np.where(mpg1['fl'] == 'c', 2.66,mpg1['price_fl'])
mpg1['price_fl'] = np.where(mpg1['fl'] == 'd', 2.23,mpg1['price_fl'])
mpg1['price_fl'] = np.where(mpg1['fl'] == 'e', 2.11,mpg1['price_fl'])
mpg1['price_fl'] = np.where(mpg1['fl'] == 'p', 2.89,mpg1['price_fl'])
mpg1['price_fl'] = np.where(mpg1['fl'] == 'r', 2.49,mpg1['price_fl'])
mpg1
```

```python
mpg2 = pd.merge(mpg,mpg1,on='fl')
mpg2
```

```python
a= mpg2.groupby(['manufacturer','class']).mean()
```

```python
mpg2.groupby('manufacturer')[['displ','price_fl']].agg(['count','max','min','mean'])
```

```python
top5 = mpg2.groupby('manufacturer')['hwy'].sum().sort_values(ascending =False).head().index
top5
```
